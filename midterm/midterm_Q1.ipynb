# Analysis -- 1
- **To get a person's email ID which has send the maximum number of email**
- Also the top 10 email-ID which are most frequently used to send emails under the key _**From**_
- Also the top 10 email -ID which have received more number of emails under the key _**To, Cc, Bcc**_

```python
#Making the frequency distribution for the data "From" "To" "Cc" "Bcc"
FromFreq = FreqDist(fromlist)
ToFreq = FreqDist(tolist)
CcFreq = FreqDist(cclist)
BccFreq = FreqDist(bcclist)
```

```python
#The total number of email taken from the key "From" and its unique count
print(FromFreq.B())
print(FromFreq.N())
print(FromFreq.most_common(10))
```

```python
#The total number of email taken from the key "To" and its unique count
print(ToFreq.B())
print(ToFreq.N())
print(ToFreq.most_common(10)[:])
```

```python
#The total number of email taken from the key "Cc" and its unique count
print(CcFreq.B())
print(CcFreq.N())
print(CcFreq.most_common(10))
```

```python
#The total number of email taken from the key "Bcc" and its unique count
print(BccFreq.B())
print(BccFreq.N())
print(BccFreq.most_common(10))
```

```python
#Sorting the dictionery with highest frequency in the top and displaying it in the attractive format along with headings
fromcol1 = ["*From* - Email - ID"]
rankcol1 = ["Rank"]
freqcol1 = ["Frequency"]
rank = 0 
print ("*From* - Email - ID".ljust(30)  +  "\t" + "Rank" +  "\t" + "Frequency")
for w in sorted(FromFreq, key=FromFreq.get, reverse=True)[:10]:
    rank = rank + 1
    print (repr(w).ljust(30)  +  "\t" + repr(rank) +  "\t" + repr(FromFreq[w]))
    fromcol1.append(w)
    rankcol1.append(rank)
    freqcol1.append(FromFreq[w])
```

```python
#Sorting the dictionery with highest frequency in the top and displaying it in the attractive format along with headings
tocol2 = ["*To* - Email - ID"]
rankcol2 = ["Rank"]
freqcol2 = ["Frequency"]
rank = 0 
print ("*To* - Email - ID".ljust(30)  +  "\t" + "Rank" +  "\t" + "Frequency")
for w in sorted(ToFreq, key=ToFreq.get, reverse=True)[:11]:
    if w == None:
        continue
    rank = rank + 1
    print (repr(w).ljust(30)  +  "\t" + repr(rank) +  "\t" + repr(ToFreq[w]))
    tocol2.append(w)
    rankcol2.append(rank)
    freqcol2.append(ToFreq[w])
```

```python
#Sorting the dictionery with highest frequency in the top and displaying it in the attractive format along with headings
cccol3 = ["*Cc* - Email - ID"]
rankcol3 = ["Rank"]
freqcol3 = ["Frequency"]
rank = 0 
print ("*Cc* - Email - ID")
for w in sorted(CcFreq, key=CcFreq.get, reverse=True)[:11]:
    if w == None:     #Avoiding the null values
        continue
    rank = rank + 1
    print (repr(w).ljust(90)  +  "\n" + "Rank - " + repr(rank) +  "\n" + "Frequency - " + repr(CcFreq[w]))
    print("\n")
    cccol3.append(w)
    rankcol3.append(rank)
    freqcol3.append(CcFreq[w])
```

# Analysis --- 2
- As an input from the analysis-1 the most frequently used email is considered
- To get on which time period this email is used frequently
- First consider on which **year** this email is used more
- Second on that particular peak year which **month** this email is used more
- Store these two time period data in a seperate **CSV file**
- Project the output by plotting the **graph**

```python
#Importing the libraries
import datetime
import dateutil.parser as dparser
```

```python
# Total number of Email taken under the key "Date"
print(len(datelist))
```

```python
#Fetching the year and month information of the most frequent user from the "datelist" 
dateyear = []
datemonth = []
for y in range(len(fromlist)):
    if fromlist[y] == fromcol1[1]:      # condition for considering the output of Analysis-1
        datetimeparse = dparser.parse(datelist[y],fuzzy=True)
        dateyear.append(datetimeparse.year)
        datemonth.append(datetimeparse.month)
```

```python
#Calcutaing the frequency of email under each year by frequency distribution
dateyearfreq = FreqDist(dateyear)
dateyearfreq
```

```python
#Making the frequency distribution in two lists for ploting them.
xyear=["Year"]
yfrequency=["Frequency"]
for a in sorted(dateyearfreq.keys()):
    xyear.append(a)
    yfrequency.append(dateyearfreq.get(a))
print(xyear)
print(yfrequency)
```

```python
#Writing the above printed output to a CSV file
#Using the local directory address path so the it can be used from any machine
import csv
if not os.path.exists('ana_[1-3]/Q1/A2'):        #Creating the master folder "OutputFiles"
    os.mkdir('ana_[1-3]/Q1/A2')
with open ('ana_[1-3]/Q1/A2/Q1_A2_R1_Email_Year.csv', 'w' , newline='') as f:
    w = csv.writer(f,delimiter = ',')
    for i in range (len(xyear)):
        w.writerow([xyear[i], yfrequency[i]])
```

```python
#removing the heading from the lists
if xyear[0] == 'Year':
    del xyear[0]
if yfrequency[0] == 'Frequency':
    del yfrequency[0]
```

```python
#Bar graph is shown using Matplotlib to project the email transaction on every year
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl
fig = plt.figure()
width = .1
ind = np.arange(len(yfrequency))
plt.bar(ind, yfrequency, width=width)
plt.xticks(ind + width / 2, xyear)
plt.title("Frequency of Email every year")
plt.xlabel('Year')
plt.ylabel('Frequency')
fig.autofmt_xdate()
plt.show()
```

```python
#To check which year has the highest email frequency
highestyear = 0
for w in sorted(dateyearfreq, key=dateyearfreq.get, reverse=True)[:1]:
    highestyear = w
print(highestyear)
```

```python
# Fetching the data for month alone on the peak transaction year and converting the digit month to string month
import datetime
datemonthpeak = []
testing=[]
for x in range(len(dateyear)):
    if dateyear[x] == highestyear:
        testing.append(datemonth[x])
        datemonthpeak.append(datetime.date(dateyear[x], datemonth[x] ,1).strftime('%b'))
print("On the peak year " + repr(highestyear) + " the overall email sent by the rank 1 email is " + repr(len(datemonthpeak))
```

```python
#The Email transaction is grouped by month wise using Frequent Distribution
datemonthpeakfreq = FreqDist(datemonthpeak)
datemonthpeakfreq
```

```python
#Ordering the month in the dictionery
import calendar
month_numeric_mapping1 = {abb: index for abb in datemonthpeakfreq 
                                    for index, long in enumerate(calendar.month_name[1:]) 
                                    if str.lower(abb) in str.lower(long)}
def getMonthNumber(item):
    return month_numeric_mapping1[item[0]]
from collections import OrderedDict
peakmonthfreq = OrderedDict(sorted(datemonthpeakfreq.items(), key=getMonthNumber))
print(peakmonthfreq)
```

```python
#Making the frequency distribution in two lists for ploting them.
addmonthyear = "Month of " + str(highestyear)
xmonth1=[addmonthyear]
yfrequency1=["Frequency"]
for a in peakmonthfreq.keys():
    xmonth1.append(a)
    yfrequency1.append(peakmonthfreq.get(a))
print(xmonth1)
print(yfrequency1)
```

# Analysis --- 3
- **To analyze the message of every email of the Rank1 person**
- Checking the most frequently used words from the subject of every email
- finding the top frequently used words excluding the stopwords and digits
- converting to lowercase and finding the frequency of each words
- store the data to the CSV file and plot the graph

```python
# The total count of email sent by the Rank1 person is
sublistrank1 = []
for y in range(len(fromlist)):
    if fromlist[y] == fromcol1[1]:      # condition for considering the output of Analysis-1
        sublistrank1.append(sublist[y])
print("The total number of email sent by Rank1 person is = " + repr(len(sublistrank1)))
```

```python
# Spliting each subject of the email in to simple words and stored into a list
subsplit = []
for x in range(len(sublistrank1)):
    subsplit.append(sublistrank1[x].split())
print(len(subsplit))
```

```python
import itertools
submerged = list(itertools.chain.from_iterable(subsplit))
print(len(submerged))
```

```python
#Using lambda function making all characters to lowercase
subject_lower = list(map(lambda x:x.lower(),submerged))
print(len(subject_lower))
```

```python
#Using lambda function filtering only the alphabet character words
subject_alpha = list(filter(lambda x: x.isalpha(), subject_lower))
print(len(subject_alpha))
```

```python
#importing the stop words from nltk and saved in list "stopwordlist"
stopwordlist = []
from nltk.corpus import stopwords
stopwordlist = stopwords.words('english')
```

```python
#Using a fuction removing the stop words from "subject_alpha"
def func():
    subject_nostopwords = []
    for k in subject_alpha:
        if k not in stopwordlist:
            subject_nostopwords.append(k)
    return subject_nostopwords
print(len(func()))
```

```python
#Importing Frequency Distribution from nltk package
from nltk import FreqDist
#Calculating the frequency and stored in dictionery "subjwordfreq"
subjwordfreq = FreqDist(func())
#The total number of words taken for the analysis which will be matching the total frequency in CSV file
print(len(subjwordfreq))
```

```python
# Sorting the dictionery with highest frequency in the top and displaying it in the attractive format along with headings
wordrow = ['Word']
rankrow = ['Rank']
freqrow = ['Frequency']
rank = 0 
print ("Word".ljust(20)   +  "\t  " + "Rank".ljust(10)  +  " \t  " + "Frequency")
for w in sorted(subjwordfreq, key=subjwordfreq.get, reverse=True):
    rank = rank + 1
    print (repr(w).ljust(20)   +  "\t  " + repr(rank).ljust(10)  +  " \t  " + repr(subjwordfreq[w]))
    wordrow.append(w)
    rankrow.append(rank)
    freqrow.append(subjwordfreq[w])
```

```python
# Writing the above printed output to a CSV file 
import csv
with open ('ana_[1-3]/Q1/A3/Q1_A3_Email_Subject.csv', 'w' , newline='') as f:
    w = csv.writer(f,delimiter = ',')
    for i in range (len(wordrow)):
        w.writerow([wordrow[i], rankrow[i], freqrow[i]])
```

```python
# with the imported the data from the CSV file, a graph is drawn using matplotlib with rank and frequency of the words
import csv
import matplotlib.pyplot as plt

x = []
y = []

with open('ana_[1-3]/Q1/A3/Q1_A3_Email_Subject.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(row[1])
        y.append(row[2])
del x[0]
x = [int(i) for i in x]
del y[0]
y = [int(i) for i in y]

plt.plot(x[:10],y[:10], label="Top 10 frequent words")
plt.xlabel('Word Rank')
plt.ylabel('Word Frequency')
plt.title("Subject word frequency of Rank1 Email")
plt.legend()
plt.show()
```
