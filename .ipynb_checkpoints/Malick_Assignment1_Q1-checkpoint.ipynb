```python
#Importing nltk package
import nltk
```

```python
#Displays the list of books from Gutenberg Corpus
nltk.corpus.gutenberg.fileids()
```

```python
#Taking the totall words from the book "austen-persuasion.txt"
TotallWords_persuasion = nltk.corpus.gutenberg.words('austen-persuasion.txt')
print(TotallWords_persuasion)
```

```python
#Displays the totall word count of the book "austen-persuasion.txt"
len(TotallWords_persuasion)
```

```python
#Taking the unique set of words alone from the book "austen-persuasion.txt"
UniqueWords_persuasion = set(TotallWords_persuasion)
```

```python
#Displays the unique word count of the book "austen-persuasion.txt"
len(UniqueWords_persuasion)
```

```python
#First converting the totall words into normal list through itertool
import itertools
l = [([x] if isinstance(x,str) else x) for x in TotallWords_persuasion]
wordlist_persuasion = list(itertools.chain(*l))
len(wordlist_persuasion) 
```

```python
#Displays the words and its frequency in the Dictionery format
wordfreq_persuasion = []
for word in wordlist_persuasion:
    wordfreq_persuasion.append(wordlist_persuasion.count(word))
D =dict(zip(wordlist_persuasion, wordfreq_persuasion))
print(D)
```

```python
#Displays the top 10 most frequent words from the book "austen-persuasion.txt" --- ****Answer****
for w in sorted(D, key=D.get, reverse=True) [0:10]:
  print (repr(w) +  "       " + repr(D[w]))
```
